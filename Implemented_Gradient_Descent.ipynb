```python
import numpy as np
import pandas as pd
import sklearn
df = pd.read_csv('training_boston_x_y_train.csv')
clmn = df.columns
df["area_square"]=df[" ZN"]**2
df["room_square"]=df[" RM"]**2
df["crime_square"]=df["# CRIM"]**2
df["tax_square"]=df[" TAX"]**2
df['constant'] = 1
y_train = df[' Y']
del df[' Y']
x_train = df.values
```

```python
print(x_train.size)
```

```python
def cost(xt,yt,m):
    M = len(xt)
    m_array_size = len(m)
    total_cost = 0
    for i in range(M):
        y = yt[i]
        x = xt[i]
        total_cost += (1/M) * ((y - (m * x).sum())**2)
    return total_cost
```

```python
def step_gradient(xt,yt, learning_rate, m):
    m_slope = np.zeros(18)
    M = len(xt)
    for i in range(M):
        x = xt[i]
        y = yt[i]
        m_slope+=(-2/M)*((y-(m*x).sum())*x)
        
    m -= (learning_rate*m_slope)
    
    return m
```

```python
def gd(x,y,learning_rate,num_iterations):
    size = 18
    m_array = np.zeros(size)
    for i in range(num_iterations):
        m_array = step_gradient(x,y,learning_rate,m_array)
        print(i, " Cost: ", cost(x,y,m_array))
    return m_array
```

```python
def run():
    learning_rate = 0.018
    num_iterations = 1000
    m_array = gd(x_train,y_train,learning_rate,num_iterations)
    print(m_array)
    return m_array
```

```python
m_array = run()
```
